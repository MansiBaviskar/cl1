```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/AirQuality.csv")
```

```python
df.head()
```

Clean the data by converting 'Date' to a datetime format and handling missing values

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.fillna(0, inplace=True)
```

```python
print(df.info())
```

## Line Plots for AQI Trends

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['AQI'], color='blue', linewidth=2)
plt.title('AQI Trend Over Time')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.grid(True)
plt.show()
```

## Individual Pollutant Levels

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['PM2.5'], label='PM2.5', color='red')
plt.plot(df['Date'], df['PM10'], label='PM10', color='green')
plt.plot(df['Date'], df['CO'], label='CO', color='purple')
plt.title('Pollutant Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Pollutant Level')
plt.legend()
plt.grid(True)
plt.show()
```

## Bar Plots to Compare AQI Values

```python
# Plot a bar plot of AQI values across dates (for the first 30 days as an example)
plt.figure(figsize=(12, 6))
plt.bar(df['Date'][:30], df['AQI'][:30], color='blue')
plt.title('AQI Values Across Different Dates')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.xticks(rotation=45)
plt.show()
```

## Box Plots for AQI Distributions

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df['PM2.5'], df['PM10'], df['CO']], labels=['PM2.5', 'PM10', 'CO'])
plt.title('Distribution of AQI Values for Different Pollutants')
plt.xlabel('Pollutants')
plt.ylabel('Levels')
plt.grid(True)
plt.show()
```

## Scatter Plot for AQI vs. Pollutant Levels

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['PM2.5'], df['AQI'], color='red', label='PM2.5')
plt.scatter(df['PM10'], df['AQI'], color='green', label='PM10')
plt.scatter(df['CO'], df['AQI'], color='purple', label='CO')
plt.title('Relationship Between Pollutant Levels and AQI')
plt.xlabel('Pollutant Levels')
plt.ylabel('AQI Value')
plt.legend()
plt.grid(True)
plt.show()
```
